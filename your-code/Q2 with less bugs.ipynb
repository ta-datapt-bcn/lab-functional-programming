```python
# Define your string handling functions below
# Minimal 3 functions
def remove_punctuation(doc):
    doc = [re.sub("[^\w\w*]", " ", string) for string in doc]
    return doc

def to_lower_case(doc):
    doc = [string.lower() for string in doc]
    return doc

def strip_html_tags(doc): 
    doc = [re.sub("<.*>", " ", string) for string in doc]
    return doc

def remove_unicode(doc):
    pattern = "[^\x00-\x7F]+|&\w+?;"
    doc = re.sub(pattern,' ', str(doc))
    return doc #Couldn't check if it works but I've been trying for hours and I can't make it run or it runs but not with the expected results.
```

```python
# Define function
def get_bow_from_docs(docs, stop_words=[]):
    
    # In the function, first define the variables you will use such as `corpus`, `bag_of_words`, and `term_freq`.
    
    corpus = []
    bag_of_words = []
    term_freq = []
    
    """
    Loop `docs` and read the content of each doc into a string in `corpus`.
    Remember to convert the doc content to lowercases and remove punctuation.
    """
    for doc in docs:
        file_txt = open(doc, "r", encoding="utf8")
        txt = file_txt.read()
        txt ="".join(txt)
        txt = strip_html_tags(txt)
        txt ="".join(txt)
        txt = to_lower_case(txt)
        txt ="".join(txt)
        txt = remove_unicode(txt)
        txt ="".join(txt)
        txt = remove_punctuation(txt)
        txt ="".join(txt)
        pattern = "\w+"
        txt = (" ".join(re.findall(pattern, txt)))
        corpus.append(txt)
    
    """
    Loop `corpus`. Append the terms in each doc into the `bag_of_words` array. The terms in `bag_of_words` 
    should be unique which means before adding each term you need to check if it's already added to the array.
    In addition, check if each term is in the `stop_words` array. Only append the term to `bag_of_words`
    if it is not a stop word.
    """

    for doc in corpus:
        split = doc.split()
        for word in split:
            if word not in bag_of_words and word not in stop_words:
                bag_of_words.append(word)
    
    
    """
    Loop `corpus` again. For each doc string, count the number of occurrences of each term in `bag_of_words`. 
    Create an array for each doc's term frequency and append it to `term_freq`.
    """

    for doc in corpus:
        frequency = []
        terms = doc.split()
        for word in bag_of_words:
            frequency.append(terms.count(word))
        term_freq.append(frequency)
    
    # Now return your output as an object
    return {
        "bag_of_words": bag_of_words,
        "term_freq": term_freq
    }
```
